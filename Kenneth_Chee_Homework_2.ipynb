# Introduction


In this homework, you'll be required to load in a dataset which has about 500 features. By using
Lasso ($L^1$) regression, we'll find the optimal constraint on the $L^1$ norm which gives us the best
$R^2$. Then we'll plot the results.

Recall we minimize the following on ** training data: $(x_i,y_i)$**

$$\min_{\beta} \frac{1}{N} \sum_{i=1}^N (y_i - \beta \cdot x_i)^2 + \lambda \|\beta \|_{L^1}.$$


Denoting $\beta_{\lambda}$ as the minimum of the above, we then choose $\lambda$ to maximize $R^2$ on **testing data: $(x_j,y_j)$**

$$ \max_{\lambda} 1 - \frac{\sum_{j} (y_j - \beta_{\lambda} \cdot x_j)^2}{\sum_j (y_j - \bar y)^2}$$

```python
# "Randomly" splitting the data into training and testing using the following from sklearn with test size 20%:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)
```

```python
# Create linear regression object
alphas = np.logspace(-8,-3,1000)  
from sklearn.linear_model import Lasso
scores=[]

for alpha in alphas:
    regr = Lasso(alpha=alpha)
    
    #Train model using training data sets
    regr.fit(X_train, y_train)
    scores.append(regr.score(X_test, y_test))
plt.plot(alphas,scores)
test_errors=scores
plt.title('Lasso Regression')
plt.xlabel('$\lambda$')
plt.ylabel('$R^2$')
plt.show()

# Change size of figure
plt.rcParams["figure.figsize"] = (10,10)
```

```python
np.max(scores)
```

```python
alpha_optim = alphas[np.argmax(scores)]
alpha_optim
```

```python
# Plotting the top coefficients
top_coef.plot(x='Dimension_no.',y='Coef_value',kind='bar',figsize=(10,10))
plt.title('Top Coefficients from Lasso Regression for optimal $\lambda$')
plt.xlabel('X')
plt.ylabel('Coefficient Value')
plt.show()
```

Many coefficients are zero in the Lasso Regression because the Lasso imposes a $L^1$ penalty that encourages sparsity whereby coefficients are given by the following optimization problem: $$\min_{\beta} \frac{1}{N}\sum\limits_{i=1}^N |y_i-\beta x_i|^2 +\lambda\sum\limits_{i=1}^N | \beta_i |$$ The $L^1$ penalization term $ \lambda\sum\limits_{i=1}^N | \beta_i |$ does not penalize larger values of $\beta$ more than smaller values (unlike the ridge regression with $L^2$ penalty that penalizes larger values more due to its quadratic term) and hence, instead of spreading the error out and making each coefficient as small as possible, it sets coefficients to zero, resulting in sparse coeficients.

Geometrically, the $L^1$ constrained norm has level sets which are boxes and this makes it more likely for the least squares loss function to hit the vertex of the $L^1$ level sets and have coefficients corresponding to the axis of the vertex to be zero (in contrast to the $L^2$ constrained norm which has circular level sets and thus has the constraint distributed more evenly among coefficients).

f) Repeat e) but using cross validation. Use error bars on the features which are the standard deviation of the 
coefficiens obtained above. For this problem I"ll walk you through the code. You just need to apply your optimal
$\alpha$ found above. 

```python
from sklearn.cross_validation import KFold
from sklearn import preprocessing
def run_cv_coeffs(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    kf = KFold(len(y),n_folds=5,shuffle=True)
    y_pred = y.copy()
    coeffs=[]
    # Iterate through folds
    for train_index, test_index in kf:
        
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        # Initialize a classifier with key word arguments
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
        coeffs.append(clf.coef_)
    return coeffs



scaler = preprocessing.StandardScaler()
X_scaled = X_df.as_matrix().astype(np.float)
X_scaled = scaler.fit_transform(X_df)

coeffs=run_cv_coeffs(X_scaled,np.array(y_df),Lasso,alpha=alpha_optim)
```

```python
def get_coeffs(coeffs):
    coeffs_avgd = [(coeffs[0][i] + coeffs[1][i] + coeffs[2][i] + coeffs[3][i] + coeffs[4][i])/5 for i in range(0,len(X_df.columns))]
    coeffs_std = [np.std([coeffs[0][i],coeffs[1][i],coeffs[2][i],coeffs[3][i],coeffs[4][i]]) for i in range(0,len(X_df.columns))]
    return coeffs_avgd, coeffs_std
coeffs_avg,coeffs_std=get_coeffs(coeffs)
```

```python
dfCoeffs = pd.DataFrame({'type':X_df.columns.values, 'coef':coeffs_avg, 'std':coeffs_std})
dfCoeffs = dfCoeffs[(dfCoeffs['coef']>1) |(dfCoeffs['coef']<-1) ]
plt.figure(figsize=(15,15))
dfCoeffs_sorted = dfCoeffs.sort_values(['coef'])[::-1]
yerr_vals = dfCoeffs_sorted['std'].values
dfCoeffs_sorted.plot(x='type',y='coef',kind='bar',yerr=yerr_vals,figsize=(15,15))
```
